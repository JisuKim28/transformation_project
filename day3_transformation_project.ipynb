**Standard Scaling**: 평균을 중심으로 데이터를 정렬해주는 방법

*   모두 평균 근처로 몰려 있게 해줘
*   평균을 0으로, 퍼짐 정도(표준편차)를 1로 맞추는 것


**Robust Scaling**: 극단적인 값(이상치)을 무시하고 정리하는 방법

*   이상치 무시하고, 중간 값 기준으로 정리해줘
*   가운데 값(중앙값)을 기준으로, 양쪽 25%~75% 범위(IQR)를 사용함


**Min-Max Scaling**: 값을 0부터 1 사이로 쭉~ 늘리거나 줄이는 방법

*   가장 작은 값은 0, 가장 큰 값은 1로 만들어줘!"
*   나머지 값들은 그 사이에 고르게 배치됨

```python
# 스케일링을 자동적으로 해주는 라이브러리
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
```
